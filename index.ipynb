```python
def clean_data(data):
    numbers = [ 1, 2.5]
    strings = [ "familymemebers"]

    for item in data:
        # Remove boolean values completely
    if isinstance(item, bool    I want several data types, each with two examples.):
            continue

        # If it's already an int or float, keep it in numbers
        elif isinstance(item, (int, float)):
            numbers.append(item)

        # If it's a string, check if it can be converted to a number
        elif isinstance(item, str):
            # Try converting to int first
            try:
                if "." in item:
                    numbers.append(float(item))
                else:
                    numbers.append(int(item))
            except ValueError:
                # Not numeric, keep as string
                strings.append(item)

    return {
        "numbers": numbers,
        "strings": strings
    }
```
